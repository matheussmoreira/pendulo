```python
# PÊNDULO

G = 9.8 # m/s^2
L = 1.0 # m

def sol_analitica(theta0, t):
    raiz = math.sqrt(G/L)
    return theta0 * math.cos(raiz*t)

def pendulo(theta):
    return -G/L * math.sin(theta)

def pendulo_simp(theta):
    return -G/L * theta

def T(t1, t2, v1, v2):
    div = math.fabs(v1)/(math.fabs(v1)+math.fabs(v2))
    return 2 * (t1 + div * (t2 - t1))

def T_simp():
    raiz = math.sqrt(L/G)
    return 2 * math.pi * raiz
```
